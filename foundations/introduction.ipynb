### 1.1 Imaginary and Complex Numbers ### 

Complex number systems are essential when dealing with the various sub-disciplines of physical chemistry as they provide the mathematical framework necessary for describing the wave-like properties and phenomena observed for molecular and even atomic behaviour. 

Complex numbers are defined as a number generally expressed using $z$, some complex number, where $z = a + bi$. In this definition, $a$ and $b$ are real numbers while $i$ represents the imaginary unit, defined as $i = \sqrt{-1}$ or $i^2 = -1$. 

Corollary to this, the complex number $z$ is made up of two parts: the real component $Re(z) = a$ and the imaginary component $Im(z) = ib$.

#### Operations with Complex Numbers ####
Basic mathematical operations can be done with complex numbers. 

##### **Addition and Subtraction** 
Addition and subtraction is performed by performing said operations on real and imaginary components respectively as shown from the following $z_1 = 5 + i6$ and $z_2 = 3 - i3$

Where for addition: 
$z_1 + z_2 = (5 + 3) + (6 + -3)i 
= 8 + 3i $\
or\
$z_1 - z_2 = (5 - 3) + (6 - -3)i \
= 2 + 9i$

Given that a complex number is multiplied by a constant $c = 2$ such as in the following: $cz_1$\
the constant $c$ will be distributed to the respective real numbers of both components (real and imaginary) as shown below\
$cz_1 = 2(5) + 2(6i) = 10 + 12i$

##### **Multiplication**
Multiplication of complex numbers are treated like standard binomials, where both imaginary and real components are treated as a term that makes up the binomial. Remember the property that $i = \sqrt{-1}$ and as such $i^2 = -1$. 

$z_1 z_2 = (5 + 6i) (3 - 3i) $\
$= 15 + -15i + 18i - 18i^2$  
$= 33 + 3i$

##### **Division** 
Division of complex numbers are unique as it requires the exploitation of the *complex conjugate* represented with an asterisk above the variable assigned to the complex number $z^*$. The *complex conjugate* as its name might suggest, is simply the conjugate of the complex number represented by the switching of the sign of the imaginary component of the complex number: 
$z = x + yi ; z^* = x - yi$

A complex number multiplied by its complex conjugate results in the elimination of the imaginary component and gives us a real number value:  
$zz^* = (x + yi) (x - yi)$  
$= x^2 - xyi + xyi + y^2$  
$= x^2 + y^2$ 

These properties can be exploited in determining the quotient of two complex numbers as shown below:  
##### $ \frac{1 + i}{2 + i}$ $\frac{2 - i}{2 - i}$  
##### $= \frac{3 + i}{5}$  
##### $= \frac{3}{5} + \frac{1}{5}i$

#### Complex Number Graphical Representations ####
Given that complex numbers contain both a real and imaginary component, complex numbers can be represented as a point on a two dimensional plane representing the number line of real numbers, and the imginary numbers plotted vertically normal to the real number line. 

This 2D plane is aptly named, the complex plane. The way that the complex number $z$ is visualized upon the complex plane using the polar coordinate system, where the complex number has a vector quantity $r$ and the angle $\theta$ on the $x$-axis representing the number's phase. 

Consequently, the vector $r$ and its scalar quantity can be expressed in the form of $r = (x^2 + y^2)^\frac{1}{2}$ and this is equated to the magnitude or absolute value of $z$ the complex number. 

As noted previously, the polar coordinate system is best used to represent this rather than upon the Cartesian coordinate system. 
Cartesian Expression: $z = x + yi$ 

To transform the complex number written in the Cartesian form to a polar coordinate form, one can use *Euler's Formula* to accomplish this: 

Euler's Formula: $e^{i\theta} = cos\theta + isin\theta$  
Where $x = rcos\theta$ and $y = rsin\theta$ 

Knowing this, one can derive the following expression for the cartesian - polar coordinate transform:  
$z = x + yi = rcos\theta + (rsin\theta)i$  
$= r(cos\theta + isin\theta)$  
$= r(e^{i\theta})$

$\therefore$ The cartesian - polar coordinate transformation of a complex number can be expressed using the following equation:  
$$re^{i\theta} = r(cos\theta + isin\theta)$$

```python
def to_polar(x, y): 
    """ 
    Converts a complex number in rectangular cartesian form (x, y) to polar form (r, θ).
    Parameters:
    x (float): The real part of the complex number.
    y (float): Imginary component of the complex number. 
    """
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x) 
    return print(f'{r}e^(i{theta})')

def to_cartesian(r, theta): 
    """ 
    Converts a complex number as a polar expression (r, θ) to rectangular cartesian form (x, y).
    Parameters: 
    r (float): Vector magnitude of the complex number 
    theta (float): Angle of the complex number in radians.
    """
    x = r*np.cos(theta) 
    y = r*np.sin(theta) 
    return print(f"{x} + {y}i")
```

```python
# Complex Number Graphical Representation 
def plot_complex_number(x, y):
    """ 
    Plots a complex number in rectangular form (x, y) on the complex plane.
    Parameters:
    x (float): The real part of the complex number.
    y (float): The imaginary part of the complex number.
    """
    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    
    plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='blue')
    
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    
    plt.title(f'Complex Number: {x} + {y}i')
    
    plt.show()
```

```python
plot_complex_number(3, 4)
```
